```python
# importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
data=pd.read_csv(r'Z:\UserLogin_data_analysis\UserLogin_data_final.csv')
data.head()
```

```python
data.info()
```

```python
data['Timestamp']=pd.to_datetime(data['Timestamp'])
```

```python
cnt_yr=data.groupby(by=data['Timestamp'].dt.date)['Username'].nunique().to_frame().reset_index()
cnt_yr.rename(columns={'Username':'Username_count'},inplace=True)
cnt_yr['Timestamp']=pd.to_datetime(cnt_yr['Timestamp'])
cnt_yr.head()
```

```python
data_tar=cnt_yr[cnt_yr['Username_count']>5]
data_tar.head()
```

```python
data_tar.info()
```

```python
fy1=data_tar[(data_tar['Timestamp']>=pd.to_datetime('01/04/2023'))&\
         (data_tar['Timestamp']<pd.to_datetime('01/04/2024'))]
fy2=data_tar[(data_tar['Timestamp']>=pd.to_datetime('01/04/2024'))&\
         (data_tar['Timestamp']<pd.to_datetime('01/04/2025'))]
fy3=data_tar[(data_tar['Timestamp']>=pd.to_datetime('01/04/2025'))&\
         (data_tar['Timestamp']<pd.to_datetime('01/04/2026'))]
```

```python
# Calculate shared max y-axis value across all datasets
max_y = max(
    fy1['Username_count'].max(),
    fy2['Username_count'].max(),
    fy3['Username_count'].max()
)

# Identify peak in fy1
peak = fy1.loc[fy1['Username_count'].idxmax()]
peak_date = peak['Timestamp']
peak_value = peak['Username_count']

# Plot fy3
plt.figure(figsize=(15, 12))
plt.plot(fy1['Timestamp'], fy1['Username_count'], label='Daily Unique Users', color='blue', linewidth=2)

# Annotate peak
plt.scatter(peak_date, peak_value, color='red', label='Peak', zorder=5)
plt.text(peak_date, peak_value + 1,
         f"Peak: {peak_value}\nDate: {peak_date.strftime('%Y-%m-%d')}",
         color='red', ha='center', fontsize=10)

# Title with average
avg = int(fy1['Username_count'].mean())
plt.suptitle('Financial Year: 2023 to 2024\n', color='Darkblue', fontsize=16)
plt.title(f'Average Unique Employees per Day: {avg}\nMaximum Number of Unique Employees in a Day: {peak_value}', fontsize=14)

# Labels and legend
plt.xlabel('Date')
plt.ylabel('User Count')
plt.legend()
plt.grid(True)

# Apply consistent y-axis scale and tick intervals
plt.ylim(0, max_y + 10)
plt.yticks(range(0, max_y + 11, 10))

plt.tight_layout()
plt.show()
```
